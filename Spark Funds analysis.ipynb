```python
# Import the numpy and pandas packages

import numpy as np
import pandas as pd
```

```python
# Reading Companies data
companies = pd.read_csv("https://cdn.upgrad.com/UpGrad/temp/d934844e-5182-4b58-b896-4ba2a499aa57/companies.txt", sep="\t", encoding = "ISO-8859-1")
#Handling encoding issues in 'permalink' column of companies dataframe
companies['permalink'] = companies[['permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
#Converting 'permalink' column values to lower case
companies['permalink']  = companies['permalink'].str.lower()
companies
```

```python
# Reading rounds2 data
rounds2 = pd.read_csv("C:/Users/Aditya/Documents/Assignment/Group assignment/rounds2.csv", encoding = "ISO-8859-1")
#Handling encoding issues in 'company_permalink' column in rounds2 dataframe
rounds2['company_permalink'] = rounds2[['company_permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
#Converting 'company_permalink' column values to lower case
rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()
rounds2
```

```python
#Finding how many unique companies are present in companies
unique = companies['permalink'].unique()
len(unique)
```

```python
#Finding how many unique companies are present in rounds2
uni= rounds2['company_permalink'].unique()
len(uni)
```

```python
#Checking info on companies dataset
companies.info()
```

```python
#Checking info on rounds2 dataset
rounds2.info()
```

```python
#To find if there any companies in the rounds2 file which are not present in companies? 
uni1= set(uni)
unique1 = set(unique)
any_companies= uni1.difference(unique1)
len(any_companies)
```

```python
#Merge two dataframes by renaming the 'permalink' column in 'companies' dataframe to 'company_permalink'
master_frame = rounds2.merge(companies.rename(columns={'permalink': 'company_permalink'}), how = 'left', on='company_permalink')

master_frame
```

```python
# Finding column-wise null percentage
round(100*(master_frame.isnull().sum(axis=0)/len(master_frame.index)),2)
```

```python
#Knowing percentage of missing values which have large percentage (greater than 5%) of Null values. 
100*(len(master_frame[master_frame.isnull().sum(axis=1) >5].index)/len(master_frame.index))
```

```python
#Verifying missing values
master_frame[np.isnan(master_frame['raised_amount_usd'])]
```

```python
# Dropping 'raised_amount_usd' rows with NAN values
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
#Knowing percentage of missing values
round(100*(master_frame.isnull().sum(axis=0)/len(master_frame.index)),2)
```

```python
# Dropping 'country_code' rows with NAN values
master_frame = master_frame.dropna(subset = ['country_code'])
#Knowing percentage of missing values
round(100*(master_frame.isnull().sum(axis=0)/len(master_frame.index)),2)
```

```python
# Dropping unused columns 
master_frame = master_frame.drop(['funding_round_code', 'state_code', 'region', 'city', 'founded_at'], axis=1)
#Knowing percentage of missing values
round(100*(master_frame.isnull().sum(axis=0)/len(master_frame.index)),2)
```

```python
#Dropping duplicate values
master_frame = master_frame.drop_duplicates()
master_frame.shape
```

```python
# Checking number of retained rows 
(len(master_frame.index)/ 114949)*100
```

```python
# Code for unit conversion to millions
conv = lambda x: '{:,.2f}'.format(x)
master_frame['raised_amount_usd'] = (master_frame['raised_amount_usd']/1000000).apply(conv)
#Verify that column data type changes after unit conversion
master_frame.info()
```

```python
#Convet raised_amount_usd to float
master_frame['raised_amount_usd'] = pd.to_numeric(master_frame['raised_amount_usd'], errors='coerce')
master_frame.info()
```

```python
# Finding average Values of investments for each funding types
avg_fund = master_frame.groupby('funding_round_type')
avg_fund = avg_fund['raised_amount_usd'].mean()
avg_fund = avg_fund.sort_values(ascending = False)
avg_fund
```

```python
#Finding top 9 countries which have received the highest total funding 
top9 = master_frame.loc[master_frame.funding_round_type ==  'venture']
top9 = top9.pivot_table(values = 'raised_amount_usd', index = 'country_code', aggfunc = 'sum')
top9 = top9.sort_values(by = 'raised_amount_usd', ascending= False)
top9 = top9.iloc[:9, ]
top9
```

```python
# Extracting the primary_sector from master_frame
master_frame = master_frame.join(master_frame['category_list'].str.split('|', expand=True).add_prefix('category_list_').fillna(np.nan))
master_frame = master_frame.drop(master_frame.ix[:,'category_list_1':'category_list_43'].head(0).columns, axis=1)
#Renaming column names
master_frame.rename(columns={'category_list':'main_category_list', 'category_list_0': 'category_list'}, inplace=True)
master_frame
```

```python
#Read mapping file
mapping = pd.read_csv("C:/Users/Aditya/Documents/Assignment/Group assignment/mapping.csv")
mapping.info()
```

```python
#Retrieving the rows which has 0 instead of na in 'category_list' column
mapping1 = mapping[(mapping.category_list.str.contains('0')) & (mapping.category_list!= 'Enterprise 2.0')]

#Replacing '0' with 'na' in the 'category_list' column
mapping1['category_list']= mapping1.category_list.apply(lambda x: x.replace('0', 'na'))
mapping1.shape
```

```python
#Concatinating both mapping and mapping1
#Replaced rows will be added to mapping file. It won't effect the next step even though we still have unreplaced rows in dataframe as while merging it will check for exact match invalid category names with '0'' will not be picked up as they won't match
mapping = pd.concat([mapping, mapping1], axis=0)
mapping
```

```python
#Dropping blanks column
mapping = mapping.drop(['Blanks'], axis=1)
```

```python
#Creating the new column main_sector which has value 1
mapping['main_sector'] = mapping[['Automotive & Sports', 'Cleantech / Semiconductors', 'Entertainment', 'Health', 'Manufacturing', 'News, Search and Messaging', 'Others', 'Social, Finance, Analytics, Advertising']].idxmax(axis=1)

#Taking only required columns 'category_list' & 'main_sector' from mapping df
mapping = mapping[['category_list', 'main_sector']]
mapping
```

```python
#Merging 2 dataframes 'master_frame' & 'mapping' dataframes
master_frame = pd.merge(master_frame, mapping, on=['category_list', 'category_list'])
master_frame
```
